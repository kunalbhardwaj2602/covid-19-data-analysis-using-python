```python
# droping the unn
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],inplace = True, axis = 1)
```

```python
# changing the date format using the pandas to_datetime function
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
```

```python
covid_df.head()
```

```python
# Active cases 
# Creating a col. of active cases
covid_df['Active_cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail() # shows last five values of datasets

# and we added a new col. which is active cases
```

# creating a povit table using pandas

```python
statewise = pd.pivot_table(covid_df, values = ["Confirmed","Deaths", "Cured"],
                           index = "State/UnionTerritory", aggfunc = max)
```

```python
# creating a column Recovery rate
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
# creating a column Mortality Rate
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by= "Confirmed", ascending = False)
```

```python
statewise
```

```python
#implementing some colours in povit table using cmap(matplotlib)
statewise.style.background_gradient(cmap = "cubehelix")

# This a first visual is created
```

## Top states with highest Deaths

```python
# Top states with highest deaths

Top_10_Deaths = covid_df.groupby(by = "State/UnionTerritory").max()[['Deaths','Date']].sort_values(by= ['Deaths'],ascending = False).reset_index()

fig = plt.figure(figsize =(18,5))

plt.title("Top 10 states with most Deaths " , size = 25)

ax = sns.barplot(data = Top_10_Deaths.iloc[:12] , y ="Deaths", x = "State/UnionTerritory", linewidth = 2 , edgecolor= "black")

plt.xlabel("States")
plt.ylabel("Total Death Cases")
plt.show()
```

## Top states with highest Recovered

```python
# Top states with highest Recovered

Top_10_Deaths = covid_df.groupby(by = "State/UnionTerritory").max()[['Cured','Date']].sort_values(by= ['Cured'],ascending = False).reset_index()

fig = plt.figure(figsize =(18,5))

plt.title("Top 10 states with most Recovered " , size = 25)

ax = sns.barplot(data = Top_10_Deaths.iloc[:12] , y ="Cured", x = "State/UnionTerritory", linewidth = 2 , edgecolor= "black")

plt.xlabel("States")
plt.ylabel("Total Cured Cases")
plt.show()
```
